```python
import pandas as pd
import matplotlib.pyplot as plt

#Create the dataframe itself using the json file
df = pd.read_json(r'rain.json')
```

```python
df
```

```python
print("df statistics: ")
df.describe()
```

```python
df.plot(x='Month', y='Temperature', label='Temp.')
df.plot(x='Month', y='Rainfall')
plt.show()
```

```python
plt.xticks(rotation=90)
plt.plot( df['Month'], df['Temperature'], label = 'Temperature')
plt.show()
#Here the months are overlapping, so we need to fix this
```

```python
plt.figure(figsize=(15,5))   #first number is width, 2nd is height
plt.plot( df['Month'], df['Temperature'], label = 'Temperature')
plt.show()
```

```python
#Plot Rainfall
plt.figure(figsize=(17,5)) 
plt.plot( df['Month'], df['Rainfall'], label = 'Rainfall')
plt.show()
```

```python
#Plot Both temperature and Rainfall against Month
plt.figure(figsize=(17,5)) 
plt.plot( df['Month'], df['Rainfall'], label = 'Rainfall')
plt.plot( df['Month'], df['Temperature'], label = 'Temperature')
plt.legend()   #To show us different colors according to label
plt.show()
```

```python
import seaborn as sns
```

```python
df = pd.read_csv(r'tempYearly.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x = 'Year',
               y = 'Temperature',
               data = df)
plt.show()
```

```python
#To increase width of plot
sns.set(rc={'figure.figsize':(12,6)})
sns.scatterplot(x = 'Year', y = 'Temperature', data = df)
plt.show()
```

```python
#regplot to see regression line
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x = 'Year',
               y = 'Temperature',
               data = df)
plt.show()
```

```python
#regplot to see regression line with rainfall and temp
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x = 'Rainfall',
               y = 'Temperature',
               data = df)
plt.show()
```

```python
data = pd.read_csv('tempYearly.csv')
data.head()
```

```python
#We will draw a joint plot which will get scatterplot in middle and histograms on the side
sns.jointplot(x = "Rainfall",
             y = "Temperature",
              data = data
             )
plt.show()
```

In the above jointplot:
----------------------------------

Scatterplot can show us outliers, we can imagine a regression line and see the outliers

Histograms can show us the distibution of the data

```python
sns.jointplot(x = "Rainfall",
             y = "Temperature",
              data = data,
              kind = "hex"   #Hexagram to visualize it abit better
             )
plt.show()
```

```python
sns.jointplot(x = "Rainfall",
             y = "Temperature",
              data = data,
              kind = "reg"   #Regression Line, so we can see a possible +ve correlation
             )
plt.show()
```
